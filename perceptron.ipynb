### Seleccionando conjunto de datos de entrenamiento y test

```python
X = iris["data"]
y = iris["target"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=13)
```

```python
train_test_split?
```

Dado que los modelos de scikit-learn están optimizados para funcionar con variables normalizadas o estandarizadas, resulta conveniente estandarizar nuestro conjunto de datos, para ello utilizamos un modulo de preprocesamiento existente en **Scikit Learn**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
X_train_std[:10]
```

```python
X_test_std[:10]
```

Como es de apreciar tenemos la misma información suministrada por el dataset, pero en este caso se encuentra completamente estandarizada

## Construimos nuestro perceptron

```python
from sklearn.linear_model import Perceptron
```

```python
Perceptron?
```

```python
ppn = Perceptron(max_iter=50, eta0=0.1, random_state=13)
ppn.fit(X_train_std,y_train)
```

## Realizamos las predicciones con base en nuestros datos de prueba

```python
y_pred = ppn.predict(X_test_std)
```

```python
y_pred
```

```python
print("""
    El número de aciertos: {}
    El número de errores: {}
    El porcentaje de aciertos: {}
    El porcentaje de errores: {}
""".format(
    (y_pred == y_test).sum(),
    (y_pred != y_test).sum(),
    ((y_pred == y_test).sum())/len(y_pred),
    ((y_pred != y_test).sum())/len(y_pred)
))
```
